```python
choices = adult_df.workclass.unique()
probs = adult_df.workclass.value_counts() / adult_df.workclass.count()
probs
adult_df.loc[adult_df.workclass == " ?", 'workclass'] = np.random.choice(choices,p=probs)
#print choices

choices = adult_df.occupation.unique()
probs = adult_df.occupation.value_counts() / adult_df.occupation.count()
probs
adult_df.loc[adult_df.occupation == " ?", 'occupation'] = np.random.choice(choices,p=probs)

choices = adult_df['native-country'].unique()
probs = adult_df['native-country'].value_counts() / adult_df['native-country'].count()
probs
adult_df.loc[adult_df['native-country'] == " ?", 'native-country'] = np.random.choice(choices,p=probs)



adult_df['native-country'].replace(' ?', 'unknown', inplace = True)
```
